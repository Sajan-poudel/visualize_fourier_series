```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.cm import get_cmap
```

```python
def calculate():
    plt.rcParams['figure.figsize'] = [8, 8]
    plt.rcParams.update({'font.size': 18})
    
    deltax = 0.001
    l = np.pi
    x = l * np.arange(-1+deltax,1+deltax,deltax)
    n = len(x)
    ndiv = int(np.floor(n/4))
    
    f = np.zeros_like(x)
    f[ndiv:2*ndiv] = (4/n)*np.arange(1,ndiv+1)
    f[2*ndiv:3*ndiv] = np.ones(ndiv) - (4/n)*np.arange(0,ndiv)
    
    fig, ax = plt.subplots()
    ax.plot(x,f,'-',color='k', LineWidth = 2)
    
    name = "Accent"
    cmap = get_cmap('tab10')
    colors = cmap.colors
    ax.set_prop_cycle(color=colors)
    
    a0 = np.sum(f)*deltax
    ffs = a0/2
    
    a = np.zeros(50)
    b = np.zeros(50)
    for k in range(20):
        a[k] = np.sum(f*np.cos(np.pi*(k+1)*x/l))*deltax
        b[k] = np.sum(f*np.sin(np.pi*(k+1)*x/l))*deltax
        ffs = ffs + a[k]*np.cos(np.pi*(k+1)*x/l) + b[k]*np.sin(np.pi*(k+1)*x/l)
        ax.plot(x,ffs,'-')
```

```python
calculate()
```
